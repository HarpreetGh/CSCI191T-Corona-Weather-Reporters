```python
import numpy as np 
import pandas as pd
#dataset from 
#https://www.kaggle.com/eeemonts/weatherclimate-data-covid19
df = pd.read_csv("./ozone.csv");
df
```

```python
df2 = pd.read_csv("./covid_countries.csv")
df2
```

```python
#get all the countries that have at least 10,000 Covid-19 cases from df2
ozone = df[df['Country/Region'].isin(df2.Country)]
```

```python
#find all unique country names
ozone['Country/Region'].unique()
```

```python
#find all unique country names
df2.sort_values(by=['Country']).Country.unique()
```

```python
#add missing country to ozone dataset manually since spelling differences.
ozone = ozone.append(df.loc[df['Country/Region'] == 'Korea, South'])
ozone = ozone.append(df.loc[df['Country/Region'] == 'US'])
ozone.index = range(0,len(ozone))
```

```python
#remove Province/State column
ozone = ozone.drop(columns=['Province/State'])
```

```python
ozone.info()
```

```python
#create cleaning list for combining multiple ozone readings in one country
dubs = ozone['Country/Region'].value_counts() #shows how many times a country occurs in dataframe

i = 0
cleaningList = []
for i in range(0, len(dubs)):
    if dubs[i] > 1: #if country occurs more than once
        cleaningList = cleaningList + [dubs.index[i]] #add to cleaning list
cleaningList
```

```python
#This for loop gets the cleaningList Countries and averages the multiple ozone readings

for i in cleaningList:     #gets country from cleaning list
    temp = ozone[ozone['Country/Region'] == i]  #create temporary dataframe of selected country
    ozone = ozone[ozone['Country/Region'] != i] #remove all exisiting occurances of selected country
    ozone.index = range(0, len(ozone))          #fix index
    ozone.loc[len(ozone)] = [i] + list(temp.mean().values) #add the average of mulitple readings to end of ozone dataframe
ozone
```

```python
#sorting dataframe for neatness
ozone.sort_values(by=['Country/Region']) 
ozone.index = range(0,len(ozone))
ozone.head(5)
```
